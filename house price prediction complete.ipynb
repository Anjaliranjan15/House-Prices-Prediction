```python
# One cell to check everything
def check_model_health(model, X_train, X_test, y_train, y_test, predictions):
    print("="*50)
    print("MODEL HEALTH CHECK")
    print("="*50)
    
    # Split check
    print(f"✅ Split: {len(X_train)} train, {len(X_test)} test")
    
    # Learning check
    train_r2 = r2_score(y_train, model.predict(X_train))
    test_r2 = r2_score(y_test, predictions)
    print(f"✅ Learning: Train R²={train_r2:.3f}, Test R²={test_r2:.3f}")
    
    # Sense check
    print(f"✅ Predictions: ${predictions.min():,.0f} to ${predictions.max():,.0f}")
    print(f"✅ Actual: ${y_test.min():,.0f} to ${y_test.max():,.0f}")
    
    # Error check
    mae = metrics.mean_absolute_error(y_test, predictions)
    print(f"✅ Average error: ${mae:,.0f}")
    
    # Overall status
    status = "🟢 HEALTHY" if test_r2 > 0.4 else "🟡 NEEDS WORK"
    print(f"\n{status} - Model is working!")
    
    return status

# Run the check
check_model_health(lm, X_train, X_test, y_train, y_test, predictions)
```
